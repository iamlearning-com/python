# Biểu đồ mô tả dữ liệu

Các nhà nghiên cứu trình bày bản tóm tắt của dữ liệu thông qua các mô tả bằng biểu đồ, Ví dụ như lập bảng phân phối tần suất cho một đội bóng rổ. Sau đó trình bày bằng đồ thị

## Các loại đồ thị

Bảng phân phối tần suất biểu thị quốc tịch của 4 đội bóng tham gia trận đấu bóng quốc tế. <br/>

$$ Tỷ lệ =  \frac { tần suất } { tổng } * 100 $$

| Quốc gia | Tần suất | Tỷ lệ (%) |
| ------- | -------- | ----- |
| Mỹ | 240 | 55.05 |
| Hàn Quốc | 30 | 6.88|
| Nhật Bản | 44 | 10.09 |
| Pháp | 56 | 12.84 |
| Đức | 66 | 15.14 |
| Tổng | 436 | 100 |

### Diễn dải bằng biểu đồ tròn:

Tóm tắt trực quan khi có ít tham số dữ liệu

Khuyết điểm: nếu có nhiều dữ liệu rời rạc khác nhau thì rất rối

```python
import matplotlib.pyplot as plt

# Dữ liệu
quoc_gia = ['Mỹ', 'Hàn Quốc', 'Nhật Bản', 'Pháp', 'Đức']
ty_le = [55.05, 6.88, 10.09, 12.84, 15.14]

# Kích cỡ biểu đồ
# plt.figure(figsize=(5, 6))
plt.pie(ty_le, labels=quoc_gia, autopct='%1.2f%%', startangle=90)
plt.title('Biểu đồ Phân phối Tỷ lệ của Các Quốc gia')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

quoc_gia = ['Mỹ', 'Hàn Quốc', 'Nhật Bản', 'Pháp', 'Đức']
ty_le = [55.05, 6.88, 10.09, 12.84, 15.14]

plt.bar(quoc_gia,ty_le)
plt.show()

plt.barh(quoc_gia,ty_le)
plt.show()
```
